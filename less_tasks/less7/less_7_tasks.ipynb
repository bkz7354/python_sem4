#### 1. Сгенерируйте следующие случайные распределения (число элементов в каждом распределении 100-500).

* Дискретное распределение Пуассона (воспользуйтесь np.random.poisson)
* Нормального распределения (Гаусса) (np.random.normal)

Постройте для каждого из распределений гистограммы и boxplots и проанализируйте их.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.poisson(4, size=1000)

fig, ax = plt.subplots(1,2)
fig.set_figwidth(12)
    
ax[0].hist(data, 40)
ax[1].boxplot(data)

xs = np.linspace(min(data), max(data), max(data) - min(data) + 1)
ys = np.array([4**(k)*np.exp(-4)/np.math.factorial(k) for k in xs])*1000

ax[0].plot(xs, ys, 'o')
```

```python
data = np.random.normal(size=1000)

fig, ax = plt.subplots(1,2)
fig.set_figwidth(12)
    
ax[0].hist(data, 40, density=True)
ax[1].boxplot(data)

xs = np.linspace(-max(data), max(data), 20)
ys = np.exp(-xs**2/2)/np.sqrt(2*np.pi)

ax[0].plot(xs, ys, 'o')
```

#### 2. Постройте график функции:

#### y(x) = k∙cos(x – a) + b

#### для некоторых (2-3 различных) значений параметров k, a, b

```python
ks = [1,2,3]
As = [3,1,4]
bs = [-0.1, 0.5, 0.2]

xs = np.linspace(0, 10, 100)

for k,a,b in zip(ks, As, bs):
    plt.plot(xs, k*np.cos(xs - a) + b)

plt.show()
```
