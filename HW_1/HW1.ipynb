#### 3. Алгоритмическая сложность (2 балла)

Задан массив X[0..N-1]. Определите ассимптотическую сложность алгоритма:

```python
S=X[0]+X[N-1]
for k in range(N):
    for m in range(N):
        for q in range(N):
            X[k]+=X[q]+S
```
